```python
import numpy as np
import pandas as pd
```

```python
dt=pd.read_csv("consumer_reviews.csv")
dt.head(10)
```

```python
dt["reviews_text"]= dt["reviews.title"]+ " "+ dt["reviews.text"]
```

```python
dt.drop(dt.columns.difference(['reviews.rating','reviews_text']), 1, inplace=True)
```

```python
dt.head(10)
```

```python
dt.info()
```

```python
dt = dt.dropna()
```

```python
conditions = [(dt['reviews.rating']>=3),(dt['reviews.rating']<3)]
choices = ['positive','negative']
dt['sentiment_label']= np.select(conditions, choices, default='positive')
```

```python
dt.head(10)
```

```python
dt=dt.drop(['reviews.rating'],axis=1)
```

```python
dt['sentiment_label'].value_counts()
```

```python
from sklearn.utils import resample
dt_majority = dt[dt.sentiment_label=='positive']
dt_minority = dt[dt.sentiment_label=='negative']
dt_minority_upsampled = resample(dt_minority, replace=True, n_samples=26751, random_state=123)
dt= pd.concat([dt_minority_upsampled, dt_majority])
```

```python
dt['sentiment_label'].value_counts()
```

```python
import re
def remove_special_characters(text, remove_digits=True):
  text=str.lower(text) 
  text= ''.join([i for i in text if not i.isdigit()])
  pattern=r'[^a-zA-z0-9\s]'
  text=re.sub(pattern,'',text) 
  return text
```

```python
dt['reviews_text']=dt['reviews_text'].apply(remove_special_characters)
dt.head()
```

```python
def tokenizer(text):
    return text.split()
```

```python
dt['reviews_text']=dt['reviews_text'].apply(tokenizer)
dt.head()
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
porter = SnowballStemmer("english", ignore_stopwords=False)
def stemmer(text):
    return [porter.stem(word) for word in text]
```

```python
dt['reviews_text']=dt['reviews_text'].apply(stemmer)
dt.head()
```

```python
from nltk.stem import WordNetLemmatizer 
```

```python
lemmatizer = WordNetLemmatizer()
def lemmitize_it(text):
    return [lemmatizer.lemmatize(word, pos = "a") for word in text]
```

```python
dt['reviews_text']=dt['reviews_text'].apply(lemmitize_it)
dt.head()
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
def stop_it(text):
    review = [word for word in text if not word in stop_words ] 
    return review
```

```python
dt['reviews_text']=dt['reviews_text'].apply(stop_it)
dt.head()
```

```python
dt['reviews_text'] = dt['reviews_text'].apply(lambda x: ' '.join(map(str, x)))
dt.head()
```

```python
y = dt['sentiment_label']
x = dt['reviews_text']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.2,shuffle=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
tfidf = TfidfVectorizer()
tf_idf_train = tfidf.fit_transform(x_train)
tf_idf_train = tfidf.transform(x_train)
tf_idf_test = tfidf.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf=MultinomialNB(alpha=6)
clf.fit(tf_idf_train,y_train)
y_pred_test=clf.predict(tf_idf_test)

from sklearn.metrics import accuracy_score
acc_log = accuracy_score(y_pred_test, y_test)*100
print("Test accuracy:",acc_log )
```

```python
def predict_sentiment(text):
    text=remove_special_characters(text)
    text=tokenizer(text)
    text=stemmer(text)
    text=lemmitize_it(text)
    text=stop_it(text)
    text_input = tfidf.transform(text)
    return clf.predict(text_input)
```

```python
input="Nice price for a bunch of batteries."
List=list(predict_sentiment(input))
sentiment = ''

if List.count("positive")== List.count("negative"):
    sentiment = 'neutral'
else :
    sentiment=max(set(List), key = List.count)
    
print(sentiment)
print(List)
```
